```python
import tkinter
from tkinter import *
from tkinter import simpledialog, messagebox 

class MyCalculator:
    def __init__(self, root):
        self.root = root
        self.root.geometry('450x400')
        self.l1 = Label(root, text = 'First')
        self.l2 = Label(root, text = 'Second')
        self.l3 = Label(root,  text = 'Result')
        self.e1 = Entry(root, relief = 'raised', borderwidth = 2)
        self.e2 = Entry(root, relief = 'raised', borderwidth = 2)
        self.e3 = Entry(root, relief = 'solid', borderwidth = 2)
        self.b1 = Button(root, text = 'Multiply', anchor = W, command = self.multiply)
        self.b2 = Button(root, text = 'Divide',  anchor = W, command = self.divide)
        self.b3 = Button(root, text = 'Addition',  anchor = W, command = self.add)
        self.b4 = Button(root, text = 'Substraction',  anchor = W,  command = self.substract)
        self.b5 = Button(root, text = 'Clear', command = self.clear, bg = 'green', fg = 'red')
        self.b6 = Button(root, text = 'Quit', command = self.quit , bg = 'green', fg = 'red')
        
        
        self.l1.grid(row = 0, column =1)
        self.l2.grid(row = 0, column = 2)
        self.l3.grid(row= 0, column = 3)
        self.e1.grid(row = 1, column = 1 )
        self.e2.grid(row = 1, column = 2)
        self.e3.grid(row = 1, column = 3)
        self.b1.grid(row = 2, column = 0, columnspan = 10, ipadx = 2, sticky = W)
        self.b2.grid(row = 2, column = 1, ipadx = 2, sticky = E)
        self.b3.grid(row = 2, column = 2, ipadx = 2, sticky = E)
        self.b4.grid(row = 2, column = 3, ipadx = 2, sticky = E)
        self.b5.grid(row = 4, column = 1, sticky = W, ipadx = 20)
        self.b6.grid(row = 4, column = 2, sticky = E, ipadx = 20)
    
        
    def multiply(self):
        try:
            self.e3.delete(0, END)
            num1 = int(self.e1.get())
            num2 = int(self.e2.get())
            res = num1*num2
            self.e3.insert(0, str(res))
        except Exception as e:
            messagebox.showerror('Error', 'Something went wrong!')
        
        
    def divide(self):
        try:
            self.e3.delete(0, END)
            
            num1 = int(self.e1.get())
            num2 = int(self.e2.get())
            res = num1/num2
            self.e3.insert(0, str(res))
        except Exception as e:
            messagebox.showerror('Error', 'Something went wrong!')
        
    def add(self):
        try:
            self.e3.delete(0, END)
            num1 = int(self.e1.get())
            num2 = int(self.e2.get())
            res = num1+num2
            self.e3.insert(0, str(res))
        except Exception as e:
            messagebox.showerror('Error', 'Something went wrong!')
        
    def substract(self):
        try:
            self.e3.delete(0, END)
            num1 = int(self.e1.get())
            num2 = int(self.e2.get())
            res = num1-num2
            self.e3.insert(0, str(res))
        except Exception as e:
            messagebox.showerror('Error', 'Something went wrong!')
        
    def clear(self):
        try:
            self.e1.delete(0, END)
            self.e2.delete(0, END)
            self.e3.delete(0, END)
            self.e1.focus()
        except Exception as e:
            messagebox.showerror('Error', 'Something went wrong!')
            
    def quit(self):
        answer = messagebox.askyesno('Quit', 'Are you sure to exit')
        if answer == True:
            name = simpledialog.askstring('Name', 'Please enter your name')
            if name is None or '':
                messagebox.showinfo('Goodbye', 'Goodby dear user')
            else:
                messagebox.showinfo('Goodbye', 'Goodbye dear '+ str(name))
                
            self.root.destroy()
        
           
        
root = tkinter.Tk()
myapp = MyCalculator(root)
root.mainloop()


```
